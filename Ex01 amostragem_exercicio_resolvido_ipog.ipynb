### Amostra estratificada de tamanho 150

```python
amostra_simples = delivery.sample(n=150)
amostra_simples
```

```python
amostra_simples.groupby('Produto ')[['valor']].count()
```

### Amostra estratificada proporcional de 20%

```python
amostra_estratificada_d = delivery.groupby('Produto ', group_keys=False).apply(lambda grupo: grupo.sample(frac=.2))

amostra_estratificada_d
```
